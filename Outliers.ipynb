```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = np.array([
    45, 48, 50, 52, 49, 51, 47, 46, 53, 54,
    50, 49, 48, 52, 51, 47, 46, 55, 5, 120, 130
])

df = pd.DataFrame({'value': data})

# Quartiles
Q1 = df['value'].quantile(0.25)   # 25th percentile
Q3 = df['value'].quantile(0.75)   # 75th percentile
IQR = Q3 - Q1                     # Interquartile Range

# Outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['value'])

# Add bounds as vertical lines
plt.axvline(lower_bound, linestyle='--', label='Lower Bound')
plt.axvline(upper_bound, linestyle='--', label='Upper Bound')

# Labels
plt.title("Boxplot with Lower and Upper Bounds")
plt.xlabel("Value")
plt.ylabel("Distribution")
plt.legend()

plt.show()
```
